```python
def encode_interest(interest):
    """Takes a string indicating interest and encodes it to an ordinal
    (numerical) variable."""
    x=0
    if (interest == 'Very important'):
        x = 1
        
    if (interest == 'Important'):
        x = 2
    
    if (interest == 'Somewhat important'):
        x = 3
        
    if (interest == 'Not very important'):
        x = 4
        
    if (interest == 'Not at all important'):
        x = 5

    return x  
```

```python
Seeeker_encoded = Seeeker_encoded.dropna()
```

```python
Seeeker_encoded.head()
```

```python
for column in Seeeker_encoded.columns:
    Seeeker_encoded[column] = Seeeker_encoded[column].map(encode_interest)
```

```python
Seeeker_encoded.head()
```

```python
corr_matrix = Seeeker_encoded.corr(method='spearman')

fig, ax = plt.subplots(figsize=(8, 6))

sns.heatmap(corr_matrix, cmap='RdBu', annot=True, fmt='.2f',
           vmin=-1, vmax=1)

plt.title("Correlations Between Desired Benefits")

# Add tight_layout to ensure the labels don't get cut off
plt.tight_layout()
plt.show()
```
